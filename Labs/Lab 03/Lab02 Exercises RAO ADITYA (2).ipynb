## Example 1

Evaluate the integral of $f(x) = x^4 - 2x^3 + 1$ on the interval $[0, 2]$: first symbolically, and then numerically using the trapezoidal rule.

Note, we can evaluate the integral symbolically as long as the function has an analytical expression that SymPy (or Maple, Mathematica, Wolfram Alpha...) knows how to solve. Otherwise, numerical integration is required.

```python
from sympy import *  # the symbolic math package

def f(x):
    return x**4 - 2*x**3 + 1

a = 0.0  # beginning of interval
b = 2.0  # end of interval
```

```python
# symbolic integration
xs = Symbol('xs', real=True)  # the variable of integration
integrate(f(xs), (xs, a, b))


def trapezoidal(func, bounds, n=1000):
    """Trapezoidal Rule for Integration

    Parameters
    ----------
        func (function) : Function to integrate
        bounds (tuple[float]) : Upper and Lower bounds of integral
        n (int, optional) : Slice Resolution

    Returns
    -------
        float : Result of integration
    """
    a,b = bounds

    h = (a-b)/n
    sum = 0.5*func(a) + 0.5*func(b) 
    for k in range(1, n):
        sum += func(a + k*h)
    
    return h*sum


n = 1000
print(f"Integration Symbolically: {integrate(f(xs), (xs, a, b))}")
print(f"Numerical Integration using Trapezoidal Rule: {trapezoidal(f, (a,b), n)}")
```

## Example 2

Let's have a look at Newman's gaussxw and gausswab code. Recall: 
* Use `gaussxw` for integration limits from $-1$ to $+1$,
* Use `gaussxwab` for integration limits from $a$ to $b$.
* Since the calculation of weights and points is expensive, use `gaussxw.py` if you are going to change the limits repeatedly (see textbook pages 167-168 for how).

```python
# %load gaussxw
import matplotlib.gridspec as gridspec

from pylab import *
def gaussxw(N):

    # Initial approximation to roots of the Legendre polynomial
    a = linspace(3,4*N-1,N)/(4*N+2)
    x = cos(pi*a+1/(8*N*N*tan(a)))

    # Find roots using Newton's method
    epsilon = 1e-15
    delta = 1.0
    while delta>epsilon:
        p0 = ones(N,float)
        p1 = copy(x)
        for k in range(1,N):
            p0,p1 = p1,((2*k+1)*x*p1-k*p0)/(k+1)
        dp = (N+1)*(p0-x*p1)/(1-x*x)
        dx = p1/dp
        x -= dx
        delta = max(abs(dx))

    # Calculate the weights
    w = 2*(N+1)*(N+1)/(N*N*(1-x*x)*dp*dp)

    return x,w

def gaussxwab(N,a,b):
    x,w = gaussxw(N)
    return 0.5*(b-a)*x+0.5*(b+a),0.5*(b-a)*w

x_10, w_10 = gaussxw(10)
x_100, w_100 = gaussxw(100)

fig = plt.figure(figsize=(14,4))
gs = gridspec.GridSpec(ncols=11, nrows=11, figure=fig)

n10_plot = fig.add_subplot(gs[:,:5])
n100_plot = fig.add_subplot(gs[:,6:])

n10_plot.bar(x_10, w_10, width=1/10, color='black')
n100_plot.bar(x_100, w_100, width=1/100, color='black')

n10_plot.set_xlabel("Position $x$")
n100_plot.set_xlabel("Position $x$")

n10_plot.set_ylabel("Weight $w$")
n100_plot.set_ylabel("Weight $w$")

# fig.text(.5, .05, "", ha='center')
```
